```python
from yfapi import YahooFinanceAPI, Interval
from datetime import datetime, timedelta
import pandas as pd 
```

```python
# Use data_dict['stock'] to make a df, then combine them 
# Working, days are incorrect unless in weekly format, also maybe monthly
# Use weekly, because that basically sums up all the info 


dh = YahooFinanceAPI(Interval.WEEKLY)
now = datetime(2022, 7, 13)
then = datetime(2022, 1, 13)


data_dict = dh.get_data_for_tickers(['msft', 'aapl', 'amzn'], then, now)
```

```python
df = data_dict['msft']
df1 = data_dict['aapl']
df2 = data_dict['amzn']
```

```python
# Removing the Adj Close from these downloaded tickers, and plan to join them later with the .CSV files
# Better to rename now or later?

df.drop(["Adj Close"], axis=1, inplace=True) 
```

```python
df1.drop(["Adj Close"], axis=1, inplace=True)
```

```python
df2.drop('Adj Close', axis=1, inplace=True)
```
